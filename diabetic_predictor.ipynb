```python
import kagglehub
import pandas as pd
import os
import matplotlib.pyplot as plt

# Download dataset directly from Kaggle
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")
# print("path:", path)
# Load CSV
csv_file = os.path.join(path, "diabetes.csv")
df = pd.read_csv(csv_file)

# Check it loaded
print("✅ Dataset loaded successfully!")
print("Shape:", df.shape)
print("First 5 rows:", df.head());

# Column names
print("Columns:", df.columns.tolist())

# Missing values
print("\nMissing values:")
print(df.isnull().sum())

# Statistics
# print("\nBasic stats:")
# df.describe()



# Glucose comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Chart 1 - Glucose by outcome
df.groupby('Outcome')['Glucose'].hist(
    alpha=0.6, ax=axes[0], bins=20
)
axes[0].set_title('Glucose: Diabetic vs Non-Diabetic')
axes[0].set_xlabel('Glucose Level')
axes[0].legend(['No Diabetes', 'Diabetes'])

# Chart 2 - BMI distribution
df['BMI'].hist(ax=axes[1], bins=20, color='green', alpha=0.7)
axes[1].set_title('BMI Distribution')
axes[1].set_xlabel('BMI')

plt.tight_layout()
plt.show()

```
